#### 11. Calculate the total quantity ordered and revenue generated from these orders.

Expected output:

````python
            Quantity    2978.0
            Revenue      637.0
````

`````python
# your code here

quantity = products_df['Quantity'].sum()
revenue = products_df['Revenue'].sum()

print('')
print(f"\033[0;43m Quantity: {quantity} ")
print('')
print(f"\033[0;45m Revenue: {revenue} ")
print('')
`````

#### 12. Obtain the prices of the most expensive and least expensive items ordered and print the difference.
Expected output:

````python
            Most expensive:  11.95
            Least expensive:  0.18
            Difference:  11.77
````

`````python
# your code here

most_expensive = products_df['UnitPrice'].max()
least_expensive = products_df['UnitPrice'].min()
difference = most_expensive - least_expensive

print('')
print(f"\033[0;41m Most expensive: {most_expensive} ")
print('')
print(f"\033[0;42m Least expensive: {least_expensive} ")
print('')
print(f"\033[1;46m Difference: {difference} ")
print('')
`````
