### From the fitbit dataframe, we want to visually understand:
- Average number of steps by month. Use the appropriate visualization to show the `median` `steps` by `month`.
- Is Fitbitter more active on weekend or workdays? 
- All plots must be in the same jupyter notebook cell.
- Try to create titles and label your axis. 

**Hints**:
- Use `Steps`, `Months_encoded` and `Work_or_Weekend` columns.
- Set your size figure to 12,4
- Explore `plt.sca`
- Explore `plt.xticks`
- Save your figures. 

```python
# avg steps/month

steps_month = fitbit['Steps'].groupby(fitbit['Months_encoded']).median()
steps_month
```

```python
work_vs_weekend = fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median()
work_vs_weekend
```

```python
# Your code here

fig, axs = plt.subplots(1, 2, figsize=(12, 4))
plt.suptitle('Is Fitbitter more active on weekend or workdays?', fontsize=20, fontweight='bold', color='navy', y=1.125)

plt.sca(axs[0])
steps_month.plot(kind='line', color='purple', linewidth=3)
plt.title('    Average Steps / month', fontsize=14, y=1.05, loc='left')
plt.xlabel('')
plt.ylabel('', loc='top')
plt.xticks(list(range(1,14)),['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec',''], rotation=45)
plt.axis(ymin=7000, ymax=15000)

plt.sca(axs[1])
work_vs_weekend.plot(kind='bar', color='green',alpha = 0.5)
plt.title('    Average Steps: Weekend vs. Workdays', fontsize=14, y=1.05, loc='left')
plt.xlabel('')
plt.ylabel('', loc='top')
plt.xticks(list(range(0,2)),['Weekend','Workdays'], rotation='horizontal')

# plt.savefig('weekend_vs_workdays.png', bbox_inches='tight')
plt.show()
```

### Answer

* Fitbit is more active on Workdays.

### Plot 3 scatter plots of the following features:
- Minutes Lightly Active vs Steps
- Minutes Very Active vs Steps
- Minutes Sedentary vs Steps

```python
# Your code here

fig, axs = plt.subplots(1, 3, figsize=(20, 6))

plt.sca(axs[0])
plt.scatter(fitbit['Minutes Lightly Active'], fitbit['Steps'], s=75, c='red', alpha=0.6)
plt.title('Minutes Lightly Active vs Steps', fontsize=16, fontweight='bold', color='navy', y=1.025);

plt.sca(axs[1])
plt.scatter(fitbit['Minutes Very Active'], fitbit['Steps'], s=75, c='green', alpha=0.6)
plt.title('Minutes Very Active vs Steps', fontsize=16, fontweight='bold', color='navy', y=1.025);

plt.sca(axs[2])
plt.scatter(fitbit['Minutes Sedentary'], fitbit['Steps'], s=75, c='blue', alpha=0.6)
plt.title('Minutes Sedentary vs Steps', fontsize=16, fontweight='bold', color='navy', y=1.025);

plt.show()
```
